```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms

from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.tsa.stattools import kpss
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.compat import lzip
```

```python
chill = pd.read_csv('data_csvs/everest_8000m_chill_time_series.csv')
chill.insert(loc=0, column='time', value=chill.index.to_numpy()*8)
chill.drop(columns=['local_time_of_forecast (Date Hour)'], inplace=True)
```

```python
chill['chill'].plot()
plt.title('Chill Time Series, Everest 8000m, May 13 to June 1 2024')
plt.xlabel('data point')
plt.ylabel('chill (degrees C)')
plt.show()
```

```python
# Let's see what we get from Moving Average Smoothing, perhaps it will get us a nicer-looking trend.
chill_rolling = chill['chill'].rolling(window=7)
chill_mas_mean = chill_rolling.mean()[6:]
plt.title('7-MA of Chill Time Series')
plt.xlabel('data point')
plt.ylabel('chill (degrees C)')
chill_mas_mean.plot()
plt.show()
```

```python
# The moving average of order 7 appears to give a smoother trend than what we see from the Classical Decomposition using a period of 3. 
# This makes sense because we are averaging out over the day before and the following day.
```

```python
# Compute and plot the residuals:
chill_ma_residuals = chill['chill'][6:] - chill_mas_mean
plt.title('7-MA Chill Residuals')
plt.xlabel('data point')
plt.ylabel('chill (degrees C)')
chill_ma_residuals.plot()
plt.show()
```

```python
# Use KPSS test to see if the residuals are stationary.
kpss_chill_ma_residuals = kpss(chill_ma_residuals, regression='c')
print('KPSS test: chill MA(7) residuals around a constant')
names = ['KPSS test statistic', 'p-value', 'truncation lag parameter', 'critical values']
lzip(names, kpss_chill_ma_residuals)
```

```python
# KPSS test says yes, confirm with ADF test:
adf_chill_ma_residuals = adfuller(chill_ma_residuals)
print('Augmented Dickey-Fuller test: chill MA(7) residuals')
names = ['ADF test statistic', 'p-value', '# of lags used', '# observations', 'critical values', 'maximized information criterion']
lzip(names, adf_chill_ma_residuals)
```

```python
# ADF test suggests that a unit root is not present. Thus we can posit that the data are stationary.
```

```python
# Let's examine the remaining autocorrelation:
plot_acf(chill_ma_residuals, alpha=0.05)
plt.axhline(y=0.163, color='b', linestyle='--')
plt.axhline(y=-0.163, color='b', linestyle='--')
plt.title('Autocorrelation of Chill MA(7) Residuals')
plt.xlabel('data point')
plt.ylabel('acf')
plt.show()
```

```python
# Partial autocorrelations:
plot_pacf(chill_ma_residuals, alpha=0.05)
plt.axhline(y=0.163, color='b', linestyle='--')
plt.axhline(y=-0.163, color='b', linestyle='--')
plt.title('Partial Autocorrelation of Chill MA(7) Residuals')
plt.xlabel('data point')
plt.ylabel('pacf')
plt.show()
```

```python
# In both cases, all autocorrelations are within the thresholds of significance.
# Let's also check the Ljung-Box test for autocorrelation.
acorr_ljungbox(chill_ma_residuals)
```

```python
# So the Ljung-Box test confirms: all p-values are above 0.05. Thus, the autocorrelation has been removed using a Moving Average model of order 7.
# Note that I examined orders of 2 to 7, and the orders lower than 7 did not remove all of the autocorrelation.
```

```python
# Finally, let's do a Breusch-Pagan test to see if the variance of the residuals is constant.
chill_ma_residuals_df = pd.DataFrame(chill_ma_residuals)
print('Mean of the residuals: ' + str(chill_ma_residuals_df.mean(axis=0)) + '\n')
chill_ma_residuals_df.insert(loc=0, column='index', value=chill_ma_residuals_df.index.to_numpy())
bp_fit = smf.ols('chill ~ index', data=chill_ma_residuals_df).fit()
print(bp_fit.summary())

names = ['Lagrange multiplier statistic', 'p-value', 'f-value', 'f p-value']
bp_result = sms.het_breuschpagan(bp_fit.resid, bp_fit.model.exog)
lzip(names, bp_result)
```

```python
# BP test indicates HETEROSCEDASTICITY here.
```

```python
wind = pd.read_csv('data_csvs/everest_8000m_wind_time_series.csv')
wind.insert(loc=0, column='time', value=wind.index.to_numpy()*8)
wind.drop(columns=['local_time_of_forecast (Date Hour)'], inplace=True)
```

```python
wind['wind_speed'].plot()
plt.title('Wind Speed Time Series, Everest 8000m, May 13 to June 1 2024')
plt.xlabel('data point')
plt.ylabel('wind speed (km/h)')
plt.show()
```

```python
# Let's see what we get from Moving Average Smoothing, perhaps it will get us a nicer-looking trend.
wind_rolling = wind['wind_speed'].rolling(window=7)
wind_mas_mean = wind_rolling.mean()[6:]
plt.title('7-MA of Wind Speed Time Series')
plt.xlabel('data point')
plt.ylabel('wind speed (km/h)')
wind_mas_mean.plot()
plt.show()
```

```python
# Compute and plot the residuals:
wind_ma_residuals = wind['wind_speed'][6:] - wind_mas_mean
plt.title('7-MA Wind Speed Residuals')
plt.xlabel('data point')
plt.ylabel('wind speed (km/h)')
wind_ma_residuals.plot()
plt.show()
```

```python
# Use KPSS test to see if the residuals are stationary.
kpss_wind_ma_residuals = kpss(wind_ma_residuals, regression='c')
print('KPSS test: chill MA(7) residuals around a constant')
names = ['KPSS test statistic', 'p-value', 'truncation lag parameter', 'critical values']
lzip(names, kpss_chill_ma_residuals)
```

```python
# KPSS test says yes, confirm with ADF test:
adf_wind_ma_residuals = adfuller(wind_ma_residuals)
print('Augmented Dickey-Fuller test: wind MA(7) residuals')
names = ['ADF test statistic', 'p-value', '# of lags used', '# observations', 'critical values', 'maximized information criterion']
lzip(names, adf_wind_ma_residuals)
```

```python
# ADF test suggests that a unit root is not present. Thus we can posit that the data are stationary.
```

```python
# Let's examine the remaining autocorrelation:
plot_acf(wind_ma_residuals, alpha=0.05)
plt.axhline(y=0.163, color='b', linestyle='--')
plt.axhline(y=-0.163, color='b', linestyle='--')
plt.title('Autocorrelation of Wind Speed MA(7) Residuals')
plt.xlabel('data point')
plt.ylabel('acf')
plt.show()
```

```python
# Partial autocorrelations:
plot_pacf(wind_ma_residuals, alpha=0.05)
plt.axhline(y=0.163, color='b', linestyle='--')
plt.axhline(y=-0.163, color='b', linestyle='--')
plt.title('Partial Autocorrelation of Chill MA(7) Residuals')
plt.xlabel('data point')
plt.ylabel('pacf')
plt.show()
```

```python
# Varying MA orders do not appear to remove the autocorrelation in a satisfactory way. 
# Indeed, as suggested by the autocorrelation plots from the original wind speed time series (see acf_pacf notebook), an AR model is probably better for this time series.
```

```python
# Finally, let's do a Breusch-Pagan test to see if the variance of the residuals is constant.
wind_ma_residuals_df = pd.DataFrame(wind_ma_residuals)
print('Mean of the residuals: ' + str(wind_ma_residuals_df.mean(axis=0)) + '\n')
wind_ma_residuals_df.insert(loc=0, column='index', value=wind_ma_residuals_df.index.to_numpy())
bp_fit = smf.ols('wind_speed ~ index', data=wind_ma_residuals_df).fit()
print(bp_fit.summary())

names = ['Lagrange multiplier statistic', 'p-value', 'f-value', 'f p-value']
bp_result = sms.het_breuschpagan(bp_fit.resid, bp_fit.model.exog)
lzip(names, bp_result)
```

```python
# Here, the residuals appear to be homoscedastic.
```
