```python
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.stats.diagnostic import acorr_ljungbox
```

```python
# The chill time series shows a "tailing-off" or "damping" in the ACF plot, while we don't really see it in the PACF plot.
# This suggests an AR process. The order is suggested from the PACF plot, at the largest significant lag. Thus, we're looking at p = 3 or p = 6.
# This is intuitive since each day contains three observations.

# Furthermore, we know that the time series is not stationary, but the differenced time series is. This suggests d = 1.
# So, we're looking at a Differenced Autoregressive Model.
```

```python
# So, let's try ARIMA(3,1,0) and ARIMA(6,1,0).
# Start with 3,1,0:
chill = pd.read_csv('data_csvs/everest_8000m_chill_time_series.csv')

chill_arima_3_1_0 = ARIMA(chill['chill'], order=(3,1,0))
chill_arima_3_1_0_fit = chill_arima_3_1_0.fit()

# Model results:
print(chill_arima_3_1_0_fit.summary())
```

```python
# Residuals
chill_arima_3_1_0_res = pd.DataFrame(chill_arima_3_1_0_fit.resid[1:])
chill_arima_3_1_0_res.plot()
```

```python
# Well 0 looks bad so I omitted it, since d = 1, but the rest looks very promising. Very much like white noise to me.
```

```python
plot_acf(chill_arima_3_1_0_res, alpha=0.05)
plt.axhline(y=0.16, color='b', linestyle='--')
plt.axhline(y=-0.16, color='b', linestyle='--')
plt.title('Autocorrelation of Chill ARIMA(3,1,0) Residuals')
plt.xlabel('lag')
plt.ylabel('ACF')
plt.show()
```

```python
# Well, we still seem to see significant autocorrelation at a lag of 6. This provides evidence in favor of the ARIMA(6,1,0) model.
```

```python
plot_pacf(chill_arima_3_1_0_res, alpha=0.05)
plt.axhline(y=0.16, color='b', linestyle='--')
plt.axhline(y=-0.16, color='b', linestyle='--')
plt.title('Partial Autocorrelation of Chill ARIMA(3,1,0) Residuals')
plt.xlabel('lag')
plt.ylabel('PACF')
plt.show()
```

```python
# Same idea here.
```

```python
# Let's see what the Ljung-Box test has to say:
acorr_ljungbox(chill_arima_3_1_0_res)
```

```python
# It actually appears to be acceptable given a significance level of 0.05!
```

```python
# So let's compare 6,1,0:
chill_arima_6_1_0 = ARIMA(chill['chill'], order=(6,1,0))
chill_arima_6_1_0_fit = chill_arima_6_1_0.fit()

# Model results:
print(chill_arima_6_1_0_fit.summary())
```

```python
# To determine whether p = 6 is a meaningful improvement on p = 3, let's consider the AIC and BIC values.
# Interestingly they are not entirely in agreement: for p = 6, AIC is lower while BIC is higher.
# Since BIC penalizes model complexity more heavily, this must be a case where the improvement isn't adequate according to this criterion.
# Whereas, under a less strict criterion like AIC, the improvement is considered to be good enough to justify the additional complexity.
# For my tastes, I prefer to stick with a less complex model. So, I favor the result from BIC here.
```

```python
# Residuals
chill_arima_6_1_0_res = pd.DataFrame(chill_arima_6_1_0_fit.resid[1:])
chill_arima_6_1_0_res.plot()
```

```python
# To me, the residuals look extremely similar between the 3,1,0 and 6,1,0 models. 
# So, this gives me additional justification to use p = 3 over 6.
```

```python
plot_acf(chill_arima_6_1_0_res, alpha=0.05)
plt.axhline(y=0.16, color='b', linestyle='--')
plt.axhline(y=-0.16, color='b', linestyle='--')
plt.title('Autocorrelation of Chill ARIMA(6,1,0) Residuals')
plt.xlabel('lag')
plt.ylabel('ACF')
plt.show()
```

```python
plot_pacf(chill_arima_6_1_0_res, alpha=0.05)
plt.axhline(y=0.16, color='b', linestyle='--')
plt.axhline(y=-0.16, color='b', linestyle='--')
plt.title('Partial Autocorrelation of Chill ARIMA(6,1,0) Residuals')
plt.xlabel('lag')
plt.ylabel('PACF')
plt.show()
```

```python
# Let's see what the Ljung-Box test has to say:
acorr_ljungbox(chill_arima_6_1_0_res)
```

```python
# In this case, I am convinced there is no significant autocorrelation. None of the p-values in the Ljung-Box test are even close to the significance level.
```
