```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
# Here, all data are for Mt. Everest at 8000 m, from May 13 to July 1 of 2024.
```

```python
chill = pd.read_csv('data_csvs/everest_8000m_chill_time_series.csv')
chill['chill'].plot()
plt.title('Chill Time Series, Everest 8000m, May 13 to July 1 2024')
plt.xlabel('data point')
plt.ylabel('chill (degrees C)')
```

```python
# We can see that the time series has a significant upwards trend, though it wanders a little in the middle. This makes sense, because it is going from Spring to Summer over these dates.
# There also appears to be a degree of seasonality on a daily basis, which makes sense because the chill at night will be much lower than during the day.
```

```python
# Plot the autocorrelation for chill data with significance level of 0.05
plot_acf(chill['chill'], alpha=0.05)
plt.axhline(y=0.16, color='b', linestyle='--')
plt.axhline(y=-0.16, color='b', linestyle='--')
plt.title('Autocorrelation of Chill Time Series')
plt.xlabel('lag')
plt.ylabel('ACF')
plt.show()
```

```python
# So, the autocorrelation is very significant. Here, the blue shaded range represents the 95% confidence interval as calculated by Bartlett's Formula.
# I have also plotted horizontal lines at +/- 0.16: this represents the boundary for 95% of spikes for a white noise time series of length 150 (+/- 1.96/(150^0.5)).
# There is some small evidence of seasonality at lags of 3 and six, since these stick up above the others a little.
# The acf values only fall below the critical threshold point around a lag of 19, which represents over 6 days (since 3 points are recorded per day).
```

```python
# Plot the PARTIAL autocorrelation for chill data with significance level of 0.05
plot_pacf(chill['chill'], alpha=0.05)
plt.axhline(y=0.16, color='b', linestyle='--')
plt.axhline(y=-0.16, color='b', linestyle='--')
plt.title('Partial Autocorrelation of Chill Time Series')
plt.xlabel('lag')
plt.ylabel('PACF')
plt.show()
```

```python
# When looking only at the PARTIAL autocorrelations (which isolates the relationship between y(t) and y(t-k)), we see much lower values.
# We can still see evidence of some seasonality at lags of 3 and 6, as mentioned above.

# With the tailing-off behavior in the ACF plot, the PACF plot and the critical threshold, an AR(3) or AR(6) model is suggested.
```

```python
wind = pd.read_csv('data_csvs/everest_8000m_wind_time_series.csv')
wind['wind_speed'].plot()
plt.title('Wind Speed Time Series, Everest 8000m, May 13 to July 1 2024')
plt.xlabel('data point')
plt.ylabel('wind speed (km/h)')
plt.show()
```

```python
# There may be a small negative trend but it appears to be much less pronounced than for the chill data.
# Again, daily seasonality appears to be evident. It should be noted that the spike in wind speed correlates with the drop in chill from the other time plot.
# Indeed, the correlation between wind speed and chill is about -0.54.
```

```python
# Plot the autocorrelation for wind speed data with significance level of 0.05
plot_acf(wind['wind_speed'], alpha=0.05)
plt.axhline(y=0.16, color='b', linestyle='--')
plt.axhline(y=-0.16, color='b', linestyle='--')
plt.title('Autocorrelation of Wind Speed Time Series')
plt.xlabel('lag')
plt.ylabel('ACF')
plt.show()
```

```python
# We  see a lot of autocorrelation like with the chill data, but it falls off much faster.
# Interestingly there is pretty strong evidence of seasonality with lags that are multiples of 3.
# This makes sense because that corresponds to the same time of day, so winds tend to be similar around the same time of day.
# The ACF values fall below 0.41 after a lag of about 10, or over 3 days.
```

```python
# Plot the PARTIAL autocorrelation for wind speed data with significance level of 0.05
plot_pacf(wind['wind_speed'], alpha=0.05)
plt.axhline(y=0.16, color='b', linestyle='--')
plt.axhline(y=-0.16, color='b', linestyle='--')
plt.title('Partial Autocorrelation of Wind Speed Time Series')
plt.xlabel('lag')
plt.ylabel('PACF')
plt.show()
```

```python
# Again, the partial autocorrelations are much lower. The seasonality at a lag of 3 stands out, and maybe 4 abut not much beyond that.
# The partial autocorrelations are well within the critical range by a lag of 4.

# With the tailing-off behavior in the ACF plot, the PACF plot and the critical threshold, an AR(3) model is suggested, or maybe AR(4) or even AR(6).
```
