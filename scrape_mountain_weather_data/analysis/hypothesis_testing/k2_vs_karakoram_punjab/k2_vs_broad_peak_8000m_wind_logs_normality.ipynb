```python
import math

import pandas as pd
import numpy as np
import statsmodels.api as sm
import scipy.stats as stats
import matplotlib.pyplot as plt
import pylab as py 
```

```python
k2 = pd.read_csv('k2_8000m.csv')
broad_peak = pd.read_csv('broad_peak_8047m.csv')
```

```python
k2.hist(column='wind_speed', bins=15)
```

```python
broad_peak.hist(column='wind_speed', bins=15)
```

```python
# The data appear to be skewed right at least, so it will either have to be transformed or we will have to do Mann-Whitney U test or
# some kind of non-parametric test...

# In any case, we should do Q-Q plots and Shapiro-Wilkes tests to verify this
```

```python
stats.probplot(k2['wind_speed'], dist='norm', plot=plt)
plt.title('K2 Wind Speeds Normal Q-Q Plot')
plt.xlabel('Theoretical quantiles')
plt.ylabel('Ordered Values')
plt.grid(True)
plt.show()
```

```python
stats.probplot(broad_peak['wind_speed'], dist='norm', plot=plt)
plt.title('Broad Peak Wind Speeds Normal Q-Q Plot')
plt.xlabel('Theoretical quantiles')
plt.ylabel('Ordered Values')
plt.grid(True)
plt.show()
```

```python
# Both plots show positive deviations at the low and high ends. 

# Let's try a log transform of the data
k2_wind = k2['wind_speed'].to_numpy()
# Because I am taking log of this column, I should replace 0s with small values
# In practice I doubt that the wind is ever actually 0 km/h near the mountain peak
# so filling replacing that with 1 km/h indicates small wind without skewing the
# data so dramatically:
k2_wind[k2_wind == 0] = 1
k2_log_wind = np.log(k2['wind_speed'])

stats.probplot(k2_log_wind, dist='norm', plot=plt)
plt.title('Log K2 Wind Speeds Normal Q-Q Plot')
plt.xlabel('Theoretical quantiles')
plt.ylabel('Ordered Values')
plt.grid(True)
plt.show()
```

```python
# Same for Broad Peak
broad_peak_wind = broad_peak['wind_speed'].to_numpy()
broad_peak_wind[broad_peak_wind == 0] = 1
broad_peak_log_wind = np.log(broad_peak['wind_speed'])

stats.probplot(broad_peak_log_wind, dist='norm', plot=plt)
plt.title('Log Broad Peak Wind Speeds Normal Q-Q Plot')
plt.xlabel('Theoretical quantiles')
plt.ylabel('Ordered Values')
plt.grid(True)
plt.show()
```

```python
# The data are maybe somewhat more normal now but I'm not convinced still
# well, onto Shapiro-Wilk tests
```

```python
# Shapiro-Wilk tests:
k2_sw_result = stats.shapiro(k2_log_wind)
print('K2 Shapiro-Wilk test statistic: ' + str(round(k2_sw_result.statistic, 3)))
print('K2 p-value: ' + str(k2_sw_result.pvalue))
```

```python
broad_peak_sw_result = stats.shapiro(broad_peak_log_wind)
print('Broad Peak Shapiro-Wilk test statistic: ' + str(round(broad_peak_sw_result.statistic, 3)))
print('Broad Peak p-value: ' + str(broad_peak_sw_result.pvalue))
```
