```python
import math

import pandas as pd
import numpy as np
import statsmodels.api as sm
import scipy.stats as stats
import matplotlib.pyplot as plt
import pylab as py 
```

```python
k2 = pd.read_csv('k2_8000m.csv')
broad_peak = pd.read_csv('broad_peak_8047m.csv')
```

```python
k2.hist(column='wind_speed', bins=15)
```

```python
broad_peak.hist(column='wind_speed', bins=15)
```

```python
# K2 descriptive statistics
k2_wind_speeds = k2['wind_speed']
print('K2 mean: ' + str(round(k2_wind_speeds.mean(), 1)))
print('K2 median: ' + str(k2_wind_speeds.median()))
print('K2 skew: ' + str(round(k2_wind_speeds.skew(), 3)))
print('K2 kurtosis: ' + str(round(k2_wind_speeds.kurtosis(), 3)))
```

```python
# So the median is less than the mean, consistent with positive skewness.
# Kurtosis may be viewed as a measure of "tail extremity", i.e. existing outliers for a sample. Also can think of as "tail heaviness".
# Note that pandas uses the Fischer definition of kurtosis, where kurtosis of normal = 0.0 (so 3 is already subtracted)
# Therefore we see positive kurtosis, so we see more outliers than the normal distribution.
```

```python
# Broad Peak descriptive statistics
broad_peak_wind_speeds = broad_peak['wind_speed']
print('Broad Peak mean: ' + str(round(broad_peak_wind_speeds.mean(), 1)))
print('Broad Peak median: ' + str(broad_peak_wind_speeds.median()))
print('Broad Peak skew: ' + str(round(broad_peak_wind_speeds.skew(), 3)))
print('Broad Peak kurtosis: ' + str(round(broad_peak_wind_speeds.kurtosis(), 3)))
```

```python
# The trends are quite similar for Broad Peak. Notably, the mean and median are slightly smaller than K2. We shall see if this difference
# is statsitically significant.
```

```python
# Q-Q plots:
```

```python
stats.probplot(k2_wind_speeds, dist='norm', plot=plt)
plt.title('K2 Wind Speeds Normal Q-Q Plot')
plt.xlabel('Theoretical quantiles')
plt.ylabel('Ordered Values')
plt.grid(True)
plt.show()
```

```python
stats.probplot(broad_peak_wind_speeds, dist='norm', plot=plt)
plt.title('Broad Peak Wind Speeds Normal Q-Q Plot')
plt.xlabel('Theoretical quantiles')
plt.ylabel('Ordered Values')
plt.grid(True)
plt.show()
```

```python
# Both K2 and Broad Peak show positive deviations at the high and low quartiles. We were generally expecting this, based on the
# descriptive statistics and appearance of the histograms.
```

```python
# Shapiro-Wilk tests:
k2_sw_result = stats.shapiro(k2_wind_speeds)
print('K2 Shapiro-Wilk test statistic: ' + str(round(k2_sw_result.statistic, 3)))
print('K2 p-value: ' + str(k2_sw_result.pvalue))
```

```python
broad_peak_sw_result = stats.shapiro(broad_peak_wind_speeds)
print('Broad Peak Shapiro-Wilk test statistic: ' + str(round(broad_peak_sw_result.statistic, 3)))
print('Broad Peak p-value: ' + str(broad_peak_sw_result.pvalue))
```

```python
# By the Shapiro-Wilk tests, both wind speed distributions deviate significantly from normality.
# On the other hand, we have relatively large datasets here, which tends to cause the Shapiro-Wilk
# test to give false negatives, so that should be considered. 
# In any case, combined with our other measurements, we probably should conclude that the data are not normally distributed.
# Therefore, we will need to use a nonparametric test.
```
